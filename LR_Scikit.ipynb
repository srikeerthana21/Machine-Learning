```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/nunnarilabs/ml/master/logistic_regression/insurance_data.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df['bought_insurance'].value_counts()
```

```python
x=df.drop('bought_insurance',axis='columns')
x
```

```python
y=df.bought_insurance#series
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
print("Total",len(x))
print("Train_size",len(x_train))
print("Test Size",len(x_test))
```

```python
model.fit(x_train,y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
y_predicted=model.predict(x_test)
y_predicted
```

```python
print("test score ",model.score(x_test,y_test))
print("train score ",model.score(x_train,y_train))
```
